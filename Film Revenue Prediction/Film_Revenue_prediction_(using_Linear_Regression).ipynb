# **Movie Revenue Prediction Using Linear Regression**

### **Importing Required Libraries for this project**


1.   Pandas for Data Management 
2.   MatPlotLib for plotting Graphs and configuring the plot
3.   Sklearn (sciktlearn) for implementing Linear Regression Algorithm






```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
x = DataFrame(data,columns = ['production_budget_usd'])
y = DataFrame(data,columns = ['worldwide_gross_usd'])
```

```python
plt.scatter(x,y,alpha = 0.3)
plt.title(label = "Film Cost vs Global Revenue")
plt.xlabel(xlabel = "Production Budget $")
plt.ylabel(ylabel = "worldwide_gross_usd $")
plt.ylim(bottom = 0, top = 3000000000)
plt.xlim(left = 0, right = 450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(x,y)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plt.figure(figsize =(10,6))
plt.scatter(x,y,alpha = 0.3)
plt.plot(x,regression.predict(x), color = 'red', linewidth = 4 )

plt.title(label = "Film Cost vs Global Revenue")
plt.xlabel(xlabel = "Production Budget $")
plt.ylabel(ylabel = "worldwide_gross_usd $")
plt.ylim(bottom = 0, top = 3000000000)
plt.xlim(left = 0, right = 450000000)
plt.show()
```

```python
regression.score(x,y)
```
